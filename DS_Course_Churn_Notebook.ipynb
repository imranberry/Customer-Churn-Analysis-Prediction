### 1.4 Functions

````python
# Reusable function — categorise customer tenure
def tenure_segment(months: int) -> str:
    """Return a churn-risk label based on months as a customer.

    Args:
        months: customer tenure in months
    Returns:
        String label: 'New', 'Growing', 'Loyal', or 'Champion'
    """
    if months <= 6:
        return "New (High Risk)"
    elif months <= 24:
        return "Growing (Medium Risk)"
    elif months <= 48:
        return "Loyal (Low Risk)"
    else:
        return "Champion (Very Low Risk)"

# Test with sample tenure values
sample_tenures = [1, 12, 30, 60, 72]
for t in sample_tenures:
    print(f"  Tenure {t:3d} months → {tenure_segment(t)}")
````

### 3.2 Selecting, Filtering & Sorting

````python
# Select specific columns
billing = df[["customerID", "Contract", "MonthlyCharges", "TotalCharges", "Churn"]]
print("Billing subset shape:", billing.shape)
billing.head()
````

````python
# Filter: month-to-month customers who churned
high_risk = df[
    (df["Contract"] == "Month-to-month") &
    (df["Churn"] == "Yes")
]
print(f"Month-to-month churners: {len(high_risk):,}")
print(f"Average monthly charge : ${high_risk['MonthlyCharges'].mean():.2f}")

# Sort by monthly charges descending
top_spenders = df.sort_values("MonthlyCharges", ascending=False).head(10)
print("\nTop 10 highest monthly charges:")
top_spenders[["customerID", "MonthlyCharges", "Contract", "Churn"]].reset_index(drop=True)
````

### 3.3 GroupBy & Aggregation

````python
# Churn rate by contract type — the most important business insight
churn_by_contract = (
    df.groupby("Contract")
    .agg(
        total_customers=("Churn", "count"),
        churned=("Churn", lambda x: (x == "Yes").sum()),
        avg_monthly=("MonthlyCharges", "mean"),
        avg_tenure=("tenure", "mean")
    )
    .assign(churn_rate=lambda x: x["churned"] / x["total_customers"])
    .sort_values("churn_rate", ascending=False)
    .reset_index()
)
churn_by_contract["churn_rate"] = churn_by_contract["churn_rate"].map("{:.1%}".format)
churn_by_contract["avg_monthly"] = churn_by_contract["avg_monthly"].map("${:.2f}".format)
churn_by_contract["avg_tenure"] = churn_by_contract["avg_tenure"].map("{:.1f} mo".format)
print(churn_by_contract.to_string(index=False))
````

````python
# Churn rate by payment method
churn_by_payment = (
    df.groupby("PaymentMethod")
    .apply(lambda g: pd.Series({
        "customers": len(g),
        "churn_rate": (g["Churn"] == "Yes").mean()
    }))
    .sort_values("churn_rate", ascending=False)
    .reset_index()
)
churn_by_payment["churn_rate"] = churn_by_payment["churn_rate"].map("{:.1%}".format)
print(churn_by_payment.to_string(index=False))
````

### 3.4 Data Cleaning

````python
# Fix: TotalCharges has 11 missing values — fill with MonthlyCharges for new customers
print("Before cleaning:")
print(f"  TotalCharges nulls: {df['TotalCharges'].isnull().sum()}")

df_clean = df.copy()

# For customers with tenure=0, TotalCharges should equal MonthlyCharges
df_clean["TotalCharges"] = df_clean["TotalCharges"].fillna(df_clean["MonthlyCharges"])

# Convert Churn to binary integer for modelling
df_clean["Churn_Binary"] = (df_clean["Churn"] == "Yes").astype(int)

print("After cleaning:")
print(f"  TotalCharges nulls: {df_clean['TotalCharges'].isnull().sum()}")
print(f"  Churn_Binary sample: {df_clean['Churn_Binary'].value_counts().to_dict()}")
````

##  Module 5 — Statistics & Mathematics
> **Week 3 · Session 6**

We apply statistical concepts directly to the churn dataset — computing
descriptive stats, testing hypotheses, and measuring feature-target correlations.


````python
from scipy import stats

# Separate retained vs churned customers
retained = df_clean[df_clean["Churn"] == "No"]["MonthlyCharges"]
churned  = df_clean[df_clean["Churn"] == "Yes"]["MonthlyCharges"]

print("=== Descriptive Statistics: Monthly Charges ===")
print(f"{'Metric':<18} {'Retained':>12} {'Churned':>12}")
print("-" * 44)
print(f"{'Count':<18} {len(retained):>12,} {len(churned):>12,}")
print(f"{'Mean':<18} ${retained.mean():>11.2f} ${churned.mean():>11.2f}")
print(f"{'Median':<18} ${retained.median():>11.2f} ${churned.median():>11.2f}")
print(f"{'Std Dev':<18} ${retained.std():>11.2f} ${churned.std():>11.2f}")
print(f"{'IQR':<18} ${retained.quantile(.75)-retained.quantile(.25):>11.2f} "
      f"${churned.quantile(.75)-churned.quantile(.25):>11.2f}")
````

````python
# Hypothesis test: do churned customers pay significantly more?
t_stat, p_value = stats.ttest_ind(churned, retained)

print("=== Independent Samples t-Test ===")
print(f"H₀: Mean MonthlyCharges is equal for churned and retained customers")
print(f"H₁: Mean MonthlyCharges is higher for churned customers")
print(f"\nt-statistic : {t_stat:.4f}")
print(f"p-value      : {p_value:.2e}")
print(f"\nConclusion   : {'Reject H₀ — significant difference (p < 0.05)' if p_value < 0.05 else 'Fail to reject H₀'}")
````

````python
# Point-biserial correlation: numeric feature vs binary Churn
numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]
print("=== Correlation with Churn (Point-Biserial r) ===")
for col in numeric_cols:
    r, p = stats.pointbiserialr(df_clean["Churn_Binary"], df_clean[col])
    sig = "✓ Significant" if p < 0.05 else "✗ Not significant"
    print(f"  {col:<20} r = {r:+.4f}  p = {p:.2e}  {sig}")

print("\n💡 Negative tenure correlation means longer customers are LESS likely to churn")
````

##  Module 6 — Machine Learning Foundations
> **Week 4 · Session 8**

Here we set up the full ML pipeline: clean the data, engineer features,
encode categoricals, scale numerics, and create the train/test split.
Everything in this section feeds directly into Modules 7 and 8.


````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# ── Step 1: Feature engineering ───────────────────────────────────────────
df_ml = df_clean.copy()

# New features derived from existing data
df_ml["AvgMonthlySpend"]  = df_ml["TotalCharges"] / (df_ml["tenure"] + 1)
df_ml["TenureSegment"]    = pd.cut(df_ml["tenure"],
                                    bins=[0, 6, 24, 48, 100],
                                    labels=["New", "Growing", "Loyal", "Champion"])
df_ml["IsHighValue"]      = (df_ml["MonthlyCharges"] > 70).astype(int)
df_ml["NumServices"]      = df_ml[["PhoneService","OnlineSecurity","OnlineBackup",
                                    "DeviceProtection","TechSupport",
                                    "StreamingTV","StreamingMovies"]].apply(
    lambda col: (col == "Yes").astype(int)).sum(axis=1)

print("Engineered features:")
print(df_ml[["tenure","MonthlyCharges","AvgMonthlySpend",
             "TenureSegment","IsHighValue","NumServices"]].head(5).to_string())
````

````python
# ── Step 2: Define feature sets ───────────────────────────────────────────
TARGET  = "Churn_Binary"
DROP    = ["customerID", "Churn", "Churn_Binary", "TotalCharges"]   # avoid leakage

df_model = df_ml.drop(columns=DROP)

NUMERIC_COLS = ["tenure", "MonthlyCharges", "AvgMonthlySpend",
                "SeniorCitizen", "IsHighValue", "NumServices"]

CATEGORICAL_COLS = [c for c in df_model.columns if c not in NUMERIC_COLS]
print(f"Numeric features   ({len(NUMERIC_COLS)}): {NUMERIC_COLS}")
print(f"Categorical features ({len(CATEGORICAL_COLS)}): {CATEGORICAL_COLS}")
````

````python
# ── Step 3: Build preprocessing pipeline ─────────────────────────────────
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, NUMERIC_COLS),
    ("cat", cat_pipeline, CATEGORICAL_COLS)
])

# ── Step 4: Train / Test split ────────────────────────────────────────────
X = df_model
y = df_ml[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print(f"Training set  : {X_train.shape[0]:,} rows")
print(f"Test set      : {X_test.shape[0]:,} rows")
print(f"Train churn % : {y_train.mean():.2%}")
print(f"Test churn %  : {y_test.mean():.2%}  (stratified — same ratio ✓)")
````

### 7.1 Logistic Regression — Baseline Model

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (classification_report, confusion_matrix,
                              roc_auc_score, RocCurveDisplay)
import warnings; warnings.filterwarnings("ignore")

# Build & train
lr_pipe = Pipeline([
    ("prep",  preprocessor),
    ("model", LogisticRegression(max_iter=1000, random_state=42))
])
lr_pipe.fit(X_train, y_train)

# Evaluate
y_pred_lr   = lr_pipe.predict(X_test)
y_prob_lr   = lr_pipe.predict_proba(X_test)[:, 1]
auc_lr      = roc_auc_score(y_test, y_prob_lr)

print("=== Logistic Regression ===")
print(classification_report(y_test, y_pred_lr, target_names=["Retained","Churned"]))
print(f"ROC-AUC: {auc_lr:.4f}")
````

### 7.2 Random Forest Classifier

````python
from sklearn.ensemble import RandomForestClassifier

rf_pipe = Pipeline([
    ("prep",  preprocessor),
    ("model", RandomForestClassifier(n_estimators=200, max_depth=10,
                                      min_samples_leaf=5, random_state=42))
])
rf_pipe.fit(X_train, y_train)

y_pred_rf = rf_pipe.predict(X_test)
y_prob_rf = rf_pipe.predict_proba(X_test)[:, 1]
auc_rf    = roc_auc_score(y_test, y_prob_rf)

print("=== Random Forest ===")
print(classification_report(y_test, y_pred_rf, target_names=["Retained","Churned"]))
print(f"ROC-AUC: {auc_rf:.4f}")
````

### 7.3 Gradient Boosting Classifier

````python
from sklearn.ensemble import GradientBoostingClassifier

gb_pipe = Pipeline([
    ("prep",  preprocessor),
    ("model", GradientBoostingClassifier(n_estimators=200, learning_rate=0.1,
                                          max_depth=4, random_state=42))
])
gb_pipe.fit(X_train, y_train)

y_pred_gb = gb_pipe.predict(X_test)
y_prob_gb = gb_pipe.predict_proba(X_test)[:, 1]
auc_gb    = roc_auc_score(y_test, y_prob_gb)

print("=== Gradient Boosting ===")
print(classification_report(y_test, y_pred_gb, target_names=["Retained","Churned"]))
print(f"ROC-AUC: {auc_gb:.4f}")
````

### 7.4 Model Comparison — Metrics & ROC Curves

````python
from sklearn.metrics import f1_score, precision_score, recall_score

models = {
    "Logistic Regression": (y_pred_lr, y_prob_lr),
    "Random Forest"      : (y_pred_rf, y_prob_rf),
    "Gradient Boosting"  : (y_pred_gb, y_prob_gb),
}

print(f"{'Model':<25} {'Precision':>10} {'Recall':>10} {'F1':>8} {'AUC':>8}")
print("─" * 65)
for name, (pred, prob) in models.items():
    prec = precision_score(y_test, pred)
    rec  = recall_score(y_test, pred)
    f1   = f1_score(y_test, pred)
    auc  = roc_auc_score(y_test, prob)
    print(f"{name:<25} {prec:>10.4f} {rec:>10.4f} {f1:>8.4f} {auc:>8.4f}")
````

```python
from flask import Flask, request, jsonify, render_template
import joblib, pandas as pd

app    = Flask(__name__)

@app.route("/")

    try:
        data  = request.get_json()

        pred  = int(pipe.predict(df)[0])
        prob  = float(pipe.predict_proba(df)[0][1])

            "churn_probability": round(prob, 4),
            "risk_label": "HIGH RISK" if prob > 0.5 else "LOW RISK"
        })
    except Exception as e:
        return jsonify({"error": str(e)}), 400

if __name__ == "__main__":
    app.run(debug=True)
```

##  Module 10 — SQL for Data Science
> **Weeks 7 · Sessions 13–14**

We load the churn data into an in-memory SQLite database and run the
same analyses we did with Pandas — but in SQL. This proves you can move
fluidly between the two tools.


````python
import sqlite3

# Create in-memory database and load the churn data
conn = sqlite3.connect(":memory:")
df_clean.to_sql("customers", conn, index=False, if_exists="replace")
print(f"Loaded {len(df_clean):,} rows into SQLite 'customers' table ✓")

def run_sql(query, conn=conn):
    """Execute a SQL query and return results as a DataFrame."""
    return pd.read_sql_query(query, conn)
````

### SQL Query 1 — Basic SELECT with WHERE and ORDER BY

````python
result = run_sql("""
    SELECT customerID, Contract, MonthlyCharges, tenure, Churn
    FROM   customers
    WHERE  Contract = 'Month-to-month'
      AND  MonthlyCharges > 80
      AND  Churn = 'Yes'
    ORDER BY MonthlyCharges DESC
    LIMIT 10;
""")
print(f"High-charge month-to-month churners: {len(result)} rows")
result
````

### SQL Query 2 — GROUP BY & Aggregate Functions

````python
run_sql("""
    SELECT
        Contract,
        COUNT(*)                                         AS total_customers,
        SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END)    AS churned,
        ROUND(AVG(MonthlyCharges), 2)                    AS avg_monthly,
        ROUND(AVG(tenure), 1)                            AS avg_tenure,
        ROUND(
            100.0 * SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END) / COUNT(*),
            1
        )                                                AS churn_rate_pct
    FROM   customers
    GROUP  BY Contract
    ORDER  BY churn_rate_pct DESC;
""")
````

### SQL Query 3 — CTE + Window Function

````python
run_sql("""
    WITH customer_value AS (
        SELECT
            customerID,
            Contract,
            MonthlyCharges,
            tenure,
            Churn,
            ROUND(MonthlyCharges * 12, 2) AS annual_value,
            RANK() OVER (
                PARTITION BY Contract
                ORDER BY MonthlyCharges DESC
            ) AS rank_in_contract
        FROM customers
    )
    SELECT *
    FROM   customer_value
    WHERE  rank_in_contract <= 3
    ORDER  BY Contract, rank_in_contract;
""")
````

### SQL Query 4 — Churn Risk by Payment Method

````python
run_sql("""
    SELECT
        PaymentMethod,
        COUNT(*)  AS customers,
        ROUND(100.0 * SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END) / COUNT(*), 1)
                  AS churn_rate_pct,
        ROUND(AVG(MonthlyCharges), 2)  AS avg_monthly_charge,
        CASE
            WHEN 100.0 * SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END) / COUNT(*) > 35
            THEN 'HIGH RISK'
            WHEN 100.0 * SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END) / COUNT(*) > 20
            THEN 'MEDIUM RISK'
            ELSE 'LOW RISK'
        END AS risk_tier
    FROM   customers
    GROUP  BY PaymentMethod
    ORDER  BY churn_rate_pct DESC;
""")
````
